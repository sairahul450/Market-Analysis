```python
#Import
import pandas as pd
import numpy as np
df = pd.read_csv("Dataset/marketing_data.csv")
df
```

Replacing Null values with median value

The column 'Income' has 24 null values which are replaced with median value and also has $ symblol and it is string, so I'm converting to numeric

```python
df[' Income '] = df[' Income '].str.replace('$', '')
df[' Income '] = df[' Income '].str.replace(',', '').astype('float')
```

```python
df[' Income '] = df[' Income '].fillna(df[' Income '].median())
```

Replace column name ' Income ' to 'Income'

```python
df.columns = ['ID', 'Year_Birth', 'Education', 'Marital_Status', 'Income',
       'Kidhome', 'Teenhome', 'Dt_Customer', 'Recency', 'MntWines',
       'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',
       'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
       'AcceptedCmp2', 'Response', 'Complain', 'Country']
```

Convert year to age in column 'Year_Birth'

```python
from datetime import datetime, date
def age(born):
    today = date.today()
    return today.year - born
  
df['Age'] = df['Year_Birth'].apply(age)
del df['Year_Birth']
```
